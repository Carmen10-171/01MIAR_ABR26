## 01 - Buscando el bug::

Un ingeniero sin experiencia en python desarrolló una función que intenta crear dos listas separadas, **listaint** y **listastr**, a partir de una lista **loriginal** donde se mezclan valores numéricos guardados cómo tipos *int* y *str*. En **listaint** quiere tener todo convertido a tipo *int* y en **listastr** quiere tener todo convertido a *str*. Sin embargo, al imprimir las listas, nota que algo no está funcionando.

Tú cómo experto en Python, debes corregirlo.


```python
# Your need to fix HERE !!!!
def separador(loriginal):
    # Creo mis listas con los tipos corregidos a partir de la original
    listaint = loriginal.copy()
    listastr = loriginal
    for i in range(len(loriginal)):
        # Itero sobre mis elementos
        if type(loriginal[i]) == int:
            # Si es entero, cambio el dato a tipo str en listastr
            listastr[i] = str(listastr[i])
        else:
            # Si es string, cambio el dato a tipo int en listaint
            listaint[i] = int(listaint[i])
    return listaint, listastr
```
